```python
import numpy as np
from scipy import ndimage
from skimage import morphology
import os 
import nibabel 
```

```python
input_path='/Users/benjaminhunter/OneDrive - Imperial College London/documents/NIFTI'
output_path='/Users/benjaminhunter/OneDrive - Imperial College London/documents/NIFTI'
thick =2 
```

```python
def Make_annulus(input_path, output_path, thick): 
    items = []
    for currentpath, folders, files in os.walk(input_path):
        for file in files:
            if file.endswith('Lesion.nii.gz'):
                items.append((currentpath, file))
    print(items)
    if thick is None:
        thick=1
    for i in items: 
        print('Loading file %s' %i[1])
        input_img = nibabel.load(i[0]+'/'+i[1])
        img_data = input_img.get_fdata()
        img = np.asarray(img_data)
        struct = ndimage.generate_binary_structure(3, 1) ### Generate a three dimensional bianry structure- look at erosion in mathmatical morphology in stik  
        dilatedMask1 = ndimage.binary_dilation(img, structure=struct, iterations=thick).astype(img.dtype) # Scipy dilation thickc voxels
        Img = dilatedMask1 - img
        Out_Img = nibabel.Nifti1Image(Img, affine=input_img.affine) # Convert to Nift (*nifiti img again) you might need to change the affine and apploy th original image header
        nibabel.save(Out_Img, i[0]+'/Annulus_'+i[1])# Save New Annulus  nifiti image to ouput path 
    return(print('Annulus segmentations saved to output directory'))
```

```python
 Make_annulus(input_path, output_path, thick)
```
