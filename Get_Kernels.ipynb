```python
from pydicom import filereader
import os
import pandas as pd
```

```python
result_files = []
for root, dirs, files in os.walk('U:\LUCADI\Images'):
    if files:
        result_files.append(os.path.join(root, sorted(files)[0]))
```

```python
result_files
```

```python
ID, Kernel, Vendor = [],[],[]

for i in result_files:
    file = i[17:32]
    print(file)
    image = filereader.dcmread(i, force=True)
    ID.append(file)
    Kernel.append(image.ConvolutionKernel)
    Vendor.append(image.Manufacturer)
#     print(image.PatientID, image.ConvolutionKernel, image.Manufacturer)

table = pd.DataFrame()
table = table.assign(Scan = ID, Kern = Kernel, Vend = Vendor)
```

```python
table.to_csv('U:\LUCADI\Images\Kernels.csv')
```
